По хорошему для удержания, необходимо выбрать горизонт анализа данных. Мы располагаем данными до 30 мая. В таблице регистрации у нас пользователи, которые зарегистрировались до 21 мая. Возьмем горизонт анализа 10 дней.

```python
# задаём горизонт анализа данных
horizon_days = 10
```

```python
# присоединяем sessions к users по столбцу user_id
result_raw = users.merge(sessions[['user_id', 'open_time']], on='user_id', how='left')
```

```python
# вычисляем лайфтайм для каждой сессии в днях
result_raw['lifetime'] = (result_raw['open_time'] - result_raw['reg_time']).dt.days
```

```python
# исключаем пользователей c отрицательным лайфтаймом и с Nan.
result_raw = result_raw.query('lifetime >= 0')
```

```python
# строим таблицу удержания для платящих пользователей

result_grouped_payer = result_raw.query('payer == True').pivot_table(
    index=['dt'], columns='lifetime', values='user_id', aggfunc='nunique'
)
```

```python
# строим таблицу удержания для не платящих пользователей

result_grouped_not_payer = result_raw.query('payer == False').pivot_table(
    index=['dt'], columns='lifetime', values='user_id', aggfunc='nunique'
)
```

```python
# вычисляем размеры когорт

cohort_sizes_payer = (
    result_raw.query('payer == True').groupby('dt')
    .agg({'user_id': 'nunique'})
    .rename(columns={'user_id': 'cohort_size'})
)

cohort_sizes_not_payer = (
    result_raw.query('payer == False').groupby('dt')
    .agg({'user_id': 'nunique'})
    .rename(columns={'user_id': 'cohort_size'})
)
```

```python
# объединяем размеры когорт и таблицу удержания

result_grouped_payer = cohort_sizes_payer.merge(
    result_grouped_payer, on='dt', how='left'
).fillna(0)

result_grouped_not_payer = cohort_sizes_not_payer.merge(
    result_grouped_not_payer, on='dt', how='left'
).fillna(0)
```

```python
# исключаем из результата все лайфтаймы, превышающие горизонт анализа
result_grouped_payer = result_grouped_payer[['cohort_size'] + list(range(horizon_days))]

result_grouped_not_payer = result_grouped_not_payer[['cohort_size'] + list(range(horizon_days))]
```

Рассчитаем коэффициенты удержания.

```python
# делим данные таблицы удержания на размеры когорт

retetion_payer = result_grouped_payer.div(result_grouped_payer['cohort_size'], axis=0).drop(columns=['cohort_size'])

retetion_not_payer = result_grouped_not_payer.div(result_grouped_not_payer['cohort_size'], axis=0).drop(columns=['cohort_size'])
```

**Посчитаем время, которое ученики проводят в приложении.**

`Средняя продолжительность сессии (average session length)`=`суммарная продолжительность сессий за период`/`общее количество сессий за тот же период`

```python
# добавим в таблице сессий критерий оплаты
sessions['payer'] = sessions['user_id'].isin(payments['user_id'])
```

```python
# посчитаем длину сессий в секундах
sessions['time_ses'] = (sessions['close_time'] - sessions['open_time']).dt.seconds
```

```python
# посчитаем, сколько всего времени провел в игре каждый пользователь
asl = sessions.groupby(['dt', 'payer']).agg({'time_ses':'sum', 'open_time':'count'})\
                                .rename(columns={'time_ses':'session_time', 'open_time':'session_count'})\
                                .reset_index()
```

```python
# посчитаем среднюю продолжительность сессии в минутах.
asl['asl'] = (asl['session_time'].div(asl['session_count'], axis=0)/60).round(2)
```

```python
# построим график средней продолжительности сессии по дням для платящих и не платящих пользователей
plt.figure(figsize=(15,8))
sns.set_style("whitegrid", {'grid.linestyle': '--'})

sns.lineplot(data=asl, x=asl['dt'], y=asl['asl'], hue=asl['payer'], style=asl['payer'], markers=True, \
             palette=['#00DFA2', '#FF0060'])
plt.xlabel('Дата', fontsize=15)
plt.ylabel('Средняя продолжительность сессии, мин', fontsize=15)
plt.title('График средней продолжительности сессии по дням', fontsize=20)
plt.grid(True)
plt.show()
```

**Вывод:** Платящие пользователи проводят в приложении больше времени, причём инетерс к приложению у них поддерживается приблизительно на одном уровне стабильно. Не платящие пользователи постепенно теряют интерес к приложению, график к 30 мая значильно ниже, чем в первые дни.

```python
# посчитаем конверсию в шаг
funnel_payer['cr_step'] = (100 - 100 * abs(funnel_payer['users_cnt'].pct_change()).fillna(0)).round(3)
funnel_not_payer['cr_step'] = (100 - 100 * abs(funnel_not_payer['users_cnt'].pct_change()).fillna(0)).round(3)
```

```python
# объединим таблицы, для построения графика
funnel_all = pd.concat([funnel_payer, funnel_not_payer], axis=0).reset_index(drop=True)
```

### Метрики монетизации для всей когорты

Чтобы посмотреть на монетизацию, стоит посмотреть на такие метрики, как LTV, ARPPU, ARPU, AOV, PF.  Начнём с LTV.

LTV, или Lifetime Value, — это «пожизненная ценность» клиента, то есть общая сумма денег, которую один клиент в среднем приносит компании со всех своих покупок. Посчитаем LTV с накоплением для всей когорты, чтобы посмотреть динамику прироста.

```python
# добавим данные о покупках в профили
users_pay = users[['user_id', 'reg_time', 'payer', 'dt']].merge(payments[['user_id', 'pay_time', 'amt']], on='user_id', how='left')
```

```python
# рассчитаем лайфтайм пользователя для каждой покупки
users_pay['lifetime'] = (users_pay['pay_time'] - users_pay['reg_time']).dt.days
```

```python
# посчитаем, как платили игроки в каждый лайфтайм
ltv = users_pay.pivot_table(index=['payer'], columns='lifetime', values='amt', aggfunc='sum')
ltv
```

```python
# посчитаем сумму выручки с накоплением
ltv = ltv.fillna(0).cumsum(axis=1)
ltv
```

```python
# посчитаем размер когорты
cohort_sizes = (
    users_pay.query('payer == True').groupby('payer')
    .agg({'user_id': 'nunique'})
    .rename(columns={'user_id': 'cohort_size'}))
```

```python
# объединяем размеры когорт и таблицу выручки
ltv = cohort_sizes.merge(ltv, on='payer', how='left').fillna(0)
```

```python
# считаем LTV: делим каждую «ячейку» в строке на размер когорты
ltv = ltv.div(ltv['cohort_size'], axis=0)
```

```python
ltv = ltv.drop(columns=['cohort_size'])
ltv
```

```python
ltv.T.plot(grid=True)
plt.legend()
plt.xlabel('Лайфтайм')
plt.title('LTV')
plt.show()
```

```python
plt.figure(figsize=(15, 4))
sns.heatmap(ltv, annot=True, fmt='.2f', cmap="YlGnBu")
plt.title('Тепловая карта LTV без разбивки')
plt.xlabel('Лайфтайм')
plt.show()
```

LTV за 10 дней дал прирост более, чем в 2 раза, что совсем не плохо.

**Посчитаем ARPPU.**

```python
arppu = (payments['amt'].sum()/payments['user_id'].nunique()).round(3)
print(f'ARPPU составляет: {arppu} уе')
```

Значит, один покупатель, в среднем, приносит 1,6 у.е. за период

**Посчитаем конверсию.**

```python
cr = round(payments['user_id'].nunique()/users['user_id'].nunique(), 2)
print(f'Конверсия составляет: {cr}')
```

**Посчитаем ARPU.**

```python
arpu = arppu*cr
print(f'ARPU составляет: {arpu} уе')
```

Доход с одного пользователя составляет 0,016 у.е. за период

**AOV: Средний чек = Доход за период / Количество покупок**

```python
aov = (payments['amt'].sum()/payments['amt'].count()).round(2)
print(f'AOV составляет: {aov} уе')
```

**Purchase Frequency: Частота покупок = Количество покупок / Число уникальных покупателей**

```python
pf = (payments['amt'].count()/payments['user_id'].nunique()).round()
print(f'Частота покупок составляет: {pf}')
```

**Вывод:** В среднем в игре один покупатель совершает 2 покупки со cредним чеком 0,72 уе. в среднем один покупатель приносит 1.6 уе. LTV за 10 дней дает прирост более, чем в два раза. LTV одного опкупателя за 10 дней составляет 16 уе.
